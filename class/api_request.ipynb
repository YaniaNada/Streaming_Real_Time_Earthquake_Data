API request via python:

```python
import requests
import pandas as pd

# helper function to display data in readable format
def display_data(data):
    df = pd.DataFrame(data)
    print(df)

# function to call the api
def call_api(url):
    # Perform a GET request to the API endpoint
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the data (assuming the response is JSON)
        data = response.json()
        display_data(data)
    else:
        print('Failed to retrieve data:', response.status_code)

call_api('https://jsonplaceholder.typicode.com/posts')
```
